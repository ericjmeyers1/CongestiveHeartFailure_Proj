Before any serious aggregation, let's see what measurements (labevents) are taken the most.

Side objective: we may try to reduce the size of the large labevents DF by filtering out low frequency measurements.

```python
D_LABITEMS = pd.read_csv('D_LABITEMS.csv')
```

```python
D_LABITEMS.head()
```

```python
D_LABITEMS.dtypes
```

```python
lab_measurements_test = labevents.merge(D_LABITEMS, left_index = True, on = "ITEMID", how = 'left')
```

```python
lab_measurements_test.ITEMID.value_counts().head(35)
#NOTE: big dropoff after 400k value counts (top 29 measurements)
```

```python
ITEMID_list = [51221,50971,50983,50912,50902,51006,50882,51265,50868,51301,51222,50931,51249,51279,\
51248,51250,51277,50960,50893,50970,50820,50802,50804,50821,50818,51275,51237,51274,50800]
```

```python
reduced_d_labitems = D_LABITEMS[D_LABITEMS.ITEMID.isin(ITEMID_list)]
reduced_d_labitems
#LIST OF 29 MOST COMMON LAB TESTS ON ALL PATIENTS
```

```python
reduced_labevents = labevents.merge(reduced_d_labitems, left_index = True, on = "ITEMID", how = 'inner')
```

```python
reduced_labevents.shape
#Not much smaller :( womp womp
```

```python
reduced_labevents.head()
```

```python
#CAUTION: CELL TAKES 20 MIN+ TO RUN
#for chunk in pd.read_csv('LABEVENTS.csv', chunksize = chunksize):
#    lab_group = chunk.groupby(['SUBJECT_ID', 'ITEMID', 'VALUEUOM']).agg(['min', 'mean', 'max'])['VALUENUM']
lab_group = reduced_labevents.groupby(['SUBJECT_ID', 'ITEMID']).agg(['min', 'mean', 'max', 'std', 'var'])['VALUENUM'].reset_index()
```

```python
lab_group.head(30)
```

```python
lab_group.nunique()
```

```python
lab_group2 = lab_group.merge(reduced_d_labitems, left_index = True, on = "ITEMID", how = 'left')
lab_group2.head()
```

```python
#test[test['SUBJECT_ID']==3]
```

```python
#test = lab_group2[lab_group2.SUBJECT_ID.isin(lab_group2.SUBJECT_ID.unique())][['SUBJECT_ID', 'LABEL', 'min', 'mean', 'max', 'std', 'var']]
#test.pivot(index='SUBJECT_ID', columns='LABEL')
```

```python
#labgroup_pivot = lab_group2.pivot(index='SUBJECT_ID', columns='LABEL')
labgroup_pivot = lab_group2.pivot(index='SUBJECT_ID', columns='LABEL', values=['min', 'mean', 'max', 'std', 'var'])
```

```python
labgroup_pivot.head()
```

```python
labgroup_pivot.columns = list(map(lambda x: x[1]+x[0], zip(labgroup_pivot.columns.get_level_values(0), map(str, labgroup_pivot.columns.get_level_values(1)))))
```

```python
labgroup_pivot.head(10)
```

```python
labgroup_pivot.isnull().sum()
```

```python
patients_labevents = patients.join(labgroup_pivot, how = 'left', on = 'SUBJECT_ID')
```

```python
patients_labevents.head()
```

```python
#checkpoint export
patients_labevents.to_csv('patients_labevents.csv')
```
